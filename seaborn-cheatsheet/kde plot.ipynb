## kernel density estimation - 
https://seaborn.pydata.org/generated/seaborn.kdeplot.html

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
cars = sns.load_dataset('mpg')
cars.dropna(inplace = True)
print(cars.shape)
cars.head()
```

# <font color='darkblue'> Univariate kde plot 

### 1. simple kde plot with respect to one feature

```python
sns.set_style('whitegrid')
sns.kdeplot(cars.horsepower)
```

### 2. Increasing the x bins 

```python
sns.kdeplot(cars.horsepower,bw_adjust = .2)
plt.box(False)
```

### 3. With respect to differnt labels in another feature( origin)

```python
sns.kdeplot(x = cars.horsepower, data = cars, hue = 'origin')
plt.box(False)
plt.tight_layout()
```

### 4. Adding filling 

```python
sns.kdeplot(cars.horsepower, shade= True)
plt.box(False)
```

### 5. Getting Cumulative density curve

```python
sns.kdeplot(cars.horsepower, cumulative = True)
plt.box(False)
```

### 6. Adding hue with some additional aesthetics 

```python
sns.set_style('whitegrid')
sns.kdeplot(
    data = cars,
    fill= True, alpha = 0.5,linewidth = 0, 
    x = 'acceleration',hue = 'cylinders',palette = 'crest'
)
plt.box(False)
```

# <font color='darkblue'> Bivariate kdeplot 

### 1. Simple bivariate kde plot 

```python
sns.set()
sns.kdeplot(x = cars.horsepower,y = cars.mpg,data = cars)
```

### 2. Adding more levels to the bivarate kde 

```python
sns.set()
sns.kdeplot(x = cars.horsepower,y = cars.mpg,data = cars, n_levels = 30)
```

### 3. Adding color bar indicator 

```python
sns.set_style('whitegrid')
sns.kdeplot(x ="horsepower",y ="mpg",data = cars,
            n_levels = 20,
            shade = True,cbar = True)
plt.box(False)
```

### 4. Comparing scatterplot with the bivarite kde

```python
sns.scatterplot(x = cars.horsepower, y = cars.mpg, color='black', alpha=0.7, s=20)
plt.xlim(0, 260)
plt.ylim(0, 55)
plt.xlabel('Horsepower', fontsize=16)
plt.ylabel('MPG', fontsize=16)
plt.tight_layout();
plt.box(False)
```

### 5. Comparing the bivariate kde(mpg-horsepower) for cylinder 4 and 8 

```python
cyl_4 = cars[cars.cylinders==4]
cyl_8 = cars[cars.cylinders==8]

plt.figure(figsize=(8,5))

sns.kdeplot(x = cyl_4.horsepower, y = cyl_4.mpg,
            cmap="Blues", shade=True, thresh = 0.01)
sns.kdeplot(x = cyl_8.horsepower, y = cyl_8.mpg,
            cmap="Reds", shade=True,thresh = 0.05 )

# labelling the axis 
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Miles per Gallon (MPG)', fontsize=14)

# How to annotate values 
plt.annotate("4 Cylinders", (105, 32), color="blue", fontsize=12, fontweight='bold')
plt.annotate("8 Cylinders", (190, 18), color="red", fontsize=12, fontweight='bold');
plt.box(False)
```

###  6. Viewing kdeplots in the pariplot 

```python
sns.pairplot(cars[cars.cylinders.isin([4,6,8])], 
             hue='cylinders', 
             vars=['mpg', 'horsepower', 'acceleration'], 
             palette='viridis'
            );
```
