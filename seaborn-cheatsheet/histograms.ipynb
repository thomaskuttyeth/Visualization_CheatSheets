```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
import pandas as pd 
```

## Penguins Dataset

```python
penguins = sns.load_dataset('penguins')
penguins.dropna(inplace = True)
penguins.head()
```

```python
sns.set_style('whitegrid')
```

## Basic histogram

```python
sns.histplot(x = "flipper_length_mm", 
             data = penguins, kde = True, bins= 20)
plt.box(False)
```

## Adding density to the y axis 

```python
sns.histplot(x = "flipper_length_mm", 
             data = penguins, kde = True,stat = 'density')
plt.box(False)
```

```python
sns.histplot(x = "flipper_length_mm", 
             data = penguins, kde = True,stat = 'probability')
plt.box(False)
```

```python
sns.histplot(x = "flipper_length_mm", 
             data = penguins, kde = True,stat = 'probability', element = 'step',
            cumulative = True, fill = False)
plt.box(False)
```

## Splitting based on categorical feature 

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue = 'species')
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins,
             hue = 'species', element = 'step')
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins,
             hue = 'species', element = 'poly')
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins,
             hue = 'species', multiple = 'layer')
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins,
             hue = 'species', multiple = 'stack')
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins,
             hue = 'species', multiple = 'fill')
```

## Bivariate Histogram 

```python
sns.histplot(x = 'flipper_length_mm',y = 'bill_length_mm',
             data = penguins,cbar = True
            )
```

```python
sns.histplot(x = 'flipper_length_mm',y = 'bill_length_mm',
             data = penguins, hue = 'species'
            )
```

```python
sns.histplot(x = 'flipper_length_mm',y = 'species',
             data = penguins, hue = 'species'
            )
```

## Styling 

```python
sns.histplot(x = 'species' , hue = 'sex', multiple = 'dodge', data = penguins,
            shrink = 0.8)
```

```python
fig = sns.histplot(x = 'species', hue = 'sex', multiple = 'dodge', 
            data = penguins, shrink = 0.8, palette = 'bone', alpha = 0.9)
```
